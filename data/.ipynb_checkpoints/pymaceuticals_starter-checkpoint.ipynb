```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"


# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)


# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_drug_df, clinical_trial_df)
combined_data_df.head()

# Display the data table for preview
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
group_data = combined_data_df[["Drug", "Timepoint", "Tumor Volume (mm3)"]]
group_data = group_data.groupby(["Drug", "Timepoint"])
group_data.mean()

# Convert to DataFrame
group_data_df = pd.DataFrame(group_data.mean())
group_data_df = group_data_df.reset_index()


# Preview DataFrame
group_data_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
group_data_stderr = group_data.sem()
# Convert to DataFrame
group_data_stderr_df = pd.DataFrame(group_data_stderr)
group_data_stderr_df =group_data_stderr_df.reset_index()

# Preview DataFrame
group_data_stderr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
group_data_df = group_data_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
group_data_stderr_df = group_data_stderr_df.pivot(index = 'Timepoint', columns='Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
group_data_df.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots(1,1)
group_data_df.plot(yerr=group_data_stderr_df, ax=ax, marker ='o', grid=True, figsize=(10, 6))


plt.title("Tumor Volume Over Time (mean with std. error)")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")



# Save the Figure

plt.savefig('tumor_volume.png')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
group_meta = combined_data_df[["Drug", "Timepoint", "Metastatic Sites"]]
group_meta = group_meta.groupby(["Drug", "Timepoint"])
group_meta.mean()

# Convert to DataFrame
group_meta_df = pd.DataFrame(group_meta.mean())
group_meta_df = group_meta_df.reset_index()
# Preview DataFrame
group_meta_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

group_meta_stderr = group_meta.sem()

# Convert to DataFrame
group_meta_stderr_df = pd.DataFrame(group_meta_stderr)
group_meta_stderr_df =group_meta_stderr_df.reset_index()

# Preview DataFrame
group_meta_stderr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
group_meta_df = group_meta_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
group_meta_stderr_df = group_meta_stderr_df.pivot(index = 'Timepoint', columns='Drug', values = 'Metastatic Sites')

# Preview that Reformatting worked
group_meta_df.head()

```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots(1,1)
group_meta_df.plot(yerr=group_meta_stderr_df, ax=ax, marker ='o', grid=True, figsize=(10, 6))


plt.title("Metastic Sites Over Time (mean with std. error)")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Number of Sites")


# Save the Figure

plt.savefig('meta_sites.png')

```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
group_mouse = combined_data_df[["Drug", "Timepoint", "Mouse ID"]]
group_mouse = group_mouse.groupby(["Drug", "Timepoint"])

# Convert to DataFrame
group_mouse_df = pd.DataFrame(group_mouse.count())
group_mouse_df = group_mouse_df.reset_index()

# Preview DataFrame
group_mouse_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
group_mouse_df = group_mouse_df.pivot(index='Timepoint', columns='Drug', values='Mouse ID')


# Preview the Data Frame
group_mouse_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
fig, ax = plt.subplots(1,1)
group_mouse_df.plot(ax=ax, marker ='o', grid=True, figsize=(10, 6))

plt.title("Mouse Survival Rate")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Survival Rate")

# Save the Figure

plt.savefig('mouse_survival.png')

# Show the Figure
plt.show()
```
